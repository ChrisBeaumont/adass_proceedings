```python
%matplotlib inline
import os

import brewer2mpl
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import requests
import pandas as pd

mpl.rcParams['axes.color_cycle'] = brewer2mpl.get_map('Paired', 'qualitative', 12).mpl_colors[1::2] + [(0.94, 0.01, 0.50)]
mpl.rcParams['figure.figsize'] = (9,6)
mpl.rcParams['font.size'] = 14
```

```python
DEV_KEY = os.environ['ADS_DEV_KEY']
BASE_URL = 'http://adslabs.org/adsabs/api/search/'

def yearly_counts(query, ack=False):
    q = ('q', query) if not ack else ('q', 'ack:%s' % query)

    params = [q,
              ('filter', 'database:astronomy'),
              ('filter', 'property:refereed'),              
              ('rows', 200),
              ('dev_key', DEV_KEY),
              ('facet', 'year'),
              ('start', 0),
              ]
    
    r = requests.get(BASE_URL, params=params)
    r.raise_for_status()
    
    result = r.json()['results']['facets']['facet_fields']['year']
    year, ct = map(int, result[::2]), map(int, result[1::2])
    result = dict(zip(year, ct))
    result = pd.Series(result)
    return result

def results(query):
    params = [('q', query),
              ('hl', 'body'),
              ('fl', 'pubdate'),
              ('filter', 'database:astronomy'),
              ('filter', 'property:refereed'),              
              ('rows', 200),
              ('dev_key', DEV_KEY),
              ('facet', 'year'),
              ('start', 0),
              ]
    
    r = requests.get(BASE_URL, params=params)
    r.raise_for_status()

    return r.json()['results']
```

```python
def trendlines(queries, ack=False, labels=None):
    labels = labels or {}    
    for q in queries:
        v = yearly_counts(q, ack=ack)
        v.ix[2014] *= 1.5  # adjust for partial year
        plt.plot(v.index, v.values, label=labels.get(q, q), lw=4)
    plt.xlim(2000, 2014)
    plt.legend(loc='upper left', frameon=False)
    plt.title('Acknowledgement Search' if ack else 'Full Text Search')
```

```python
LANGS = ['IDL', 'Python', "C Programming Language", "R Programming Language", 'Fortran', 'MATLAB', 'Java']

trendlines(LANGS, labels={'C Programming Language': 'C/C++', 'R Programming Language': 'R'})

plt.title("Programming Language Mentions in ADS")
plt.xlabel("Year")
plt.ylabel("Refereed ADS Publications")
plt.savefig('languages.eps')
```
